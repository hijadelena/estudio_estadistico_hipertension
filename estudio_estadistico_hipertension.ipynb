```python
import pandas as pd

def medidas_centralizacion(datos):
    # Conteo (count)
    conteo = datos.count()
    
    # Media (mean)
    media = datos.mean()
    
    # Desviación estándar (std)
    desviacion_estandar = datos.std()
    
    # Valor mínimo (min)
    valor_minimo = datos.min()
    
    # Cuartil 25% (25%)
    cuartil_25 = datos.quantile(0.25)
    
    # Mediana (50%)
    mediana = datos.median()
    
    # Cuartil 75% (75%)
    cuartil_75 = datos.quantile(0.75)
    
    # Valor máximo (max)
    valor_maximo = datos.max()
    
    # Crear un índice para el DataFrame
    indice = ['Conteo', 'Media', 'Desviación Estándar', 'Valor Mínimo', 'Cuartil 25%', 'Mediana', 'Cuartil 75%', 'Valor Máximo']
    
    # Imprimir resultados
    resultados = pd.DataFrame({
        'Resultado': [conteo, media, desviacion_estandar, valor_minimo, cuartil_25, mediana, cuartil_75, valor_maximo]
    }, index=indice)
    
    print(resultados)

def medidas_dispercion(datos):
    # Rango intercuartílico (IQR)
    iqr = datos.quantile(0.75) - datos.quantile(0.25)
    
    # Mínimo absoluto (min)
    minimo_absoluto = datos.min()
    
    # Máximo absoluto (max)
    maximo_absoluto = datos.max()
    
    # Crear un índice para el DataFrame
    indice = ['Rango Intercuartílico (IQR)', 'Mínimo Absoluto', 'Máximo Absoluto']
    
    # Imprimir resultados
    resultados = pd.DataFrame({
        'Resultado': [iqr, minimo_absoluto, maximo_absoluto]
    }, index=indice)
    
    print(resultados)

# Aplicar las funciones a tus datos
print("Medidas de Centralización para Grupo 1:")
medidas_centralizacion(grupo1)

print("\nMedidas de Dispersión para Grupo 1:")
medidas_dispercion(grupo1)

```

la desviación estándar relativamente pequeña, junto con el hecho de que la media y la mediana no difieren significativamente, s
sugiere que **la media de 123.425 es representativa**. Además, la amplitud de los datos no es extremadamente grande.

Para la presión sistólica media en el momento de la ingestión en el grupo 2, utilizare la fórmula del intervalo de confianza para la media


```python
import pandas as pd
from scipy import stats

HIPERTENSION = pd.read_excel(r'C:\Users\lucia\Desktop\MASTER\ESTADISTICA\Documentación estadística-20231027\HIPERTENSION.xlsx')

# Filtrar x grupo2
grupo2 = HIPERTENSION[HIPERTENSION['Grupo'] == 2]

# Seleccionar la columna de presión sistólica antes en el grupo2
presion_sistolica_antes = grupo2['Presion sistólica antes']

# la media y el error estándar de la media
media = presion_sistolica_antes.mean()
std_error = stats.sem(presion_sistolica_antes)

# Calcular el intervalo de confianza al 95% y al 99%
conf_int_95 = stats.t.interval(0.95, len(presion_sistolica_antes) - 1, loc=media, scale=std_error)
conf_int_99 = stats.t.interval(0.99, len(presion_sistolica_antes) - 1, loc=media, scale=std_error)

# Imprimir resultados
print(f"Intervalo de confianza al 95%: ({conf_int_95[0]}, {conf_int_95[1]})")
print(f"Intervalo de confianza al 99%: ({conf_int_99[0]}, {conf_int_99[1]})")
```

Estos intervalos nos indican con un 95% y 99% de confianza, dónde esperamos que se encuentre la verdadera media de la presión sistólica en la población de adultos.
Dado que el valor de 130 mm Hg está dentro de ambos intervalos de confianza, no hay evidencia concluyente para rechazar la afirmación de que la presión sistólica media en el momento de la ingestión es 130 mm Hg.

Estudiaremos la proporción de la población con una presión
sistólica inicial igual o superior a 130 mm de Hg (prehipertensión). A
partir de la muestra del fichero (tomando todos los datos de presión
sistólica antes de la toma del medicamento) obtener un intervalo de
confianza al 99% de la proporción de la población con hipertensión y
contrastar la hipótesis que el porcentaje de la población con presión
sistólica superior o igual a 130 mm de Hg es 0,30 con nivel de
significación del 5%.


```python
import statsmodels.api as sm
import pandas as pd

# Supongamos que tienes un DataFrame llamado HIPERTENSION
HIPERTENSION = pd.read_excel(r'C:\Users\lucia\Desktop\MASTER\ESTADISTICA\Documentación estadística-20231027\HIPERTENSION.xlsx')

# Separar los datos en grupo1 y grupo2
grupo1 = HIPERTENSION[HIPERTENSION['Grupo'] == 1]['Presion sistólica antes']
grupo2 = HIPERTENSION[HIPERTENSION['Grupo'] == 2]['Presion sistólica antes']

# Calcular el número total de observaciones en cada grupo
n_grupo1 = len(grupo1)
n_grupo2 = len(grupo2)

# Calcular el número de observaciones con presión sistólica >= 130 en cada grupo
x_grupo1 = (grupo1 >= 130).sum()
x_grupo2 = (grupo2 >= 130).sum()

# Convertir a enteros
x_grupo1 = int(x_grupo1)
x_grupo2 = int(x_grupo2)

# Realizar la prueba de proporciones z
z_stat, p_value = sm.stats.proportions_ztest([x_grupo1, x_grupo2], [n_grupo1, n_grupo2])

# Imprimir los resultados
print(f'Estadístico Z: {z_stat}')
print(f'p-valor: {p_value}')

# Calcular el intervalo de confianza al 99%
conf_int = sm.stats.proportion_confint([x_grupo1, x_grupo2], [n_grupo1, n_grupo2], alpha=0.01)

# Imprimir el intervalo de confianza
print(f'Intervalo de confianza al 99%: ({conf_int[0][0]:.4f}, {conf_int[1][0]:.4f})')

# Contrastar la hipótesis que el porcentaje de la población con presión sistólica >= 130 mm de Hg es 0.30
hipotesis_proporcion = 0.30
z_stat_hipotesis, p_value_hipotesis = sm.stats.proportions_ztest([x_grupo1, x_grupo2], [n_grupo1, n_grupo2], value=hipotesis_proporcion)

# Imprimir los resultados de la prueba de hipótesis
print(f'Estadístico Z para la hipótesis: {z_stat_hipotesis}')
print(f'p-valor para la hipótesis: {p_value_hipotesis}')

# resultados de la prueba de hipótesis
alpha_hipotesis = 0.05
if p_value_hipotesis < alpha_hipotesis:
    print(f'Se rechaza la hipótesis nula. Hay evidencia de que la proporción es diferente de {hipotesis_proporcion}.')
else:
    print(f'No se puede rechazar la hipótesis nula. No hay suficiente evidencia de que la proporción sea diferente de {hipotesis_proporcion}.')

```

Se puede sugerir que la proporción de la población con presión sistólica igual
o superior a 130 mm de Hg es diferente de 0.3.
los resultados indican que hay evidencia estadística significativa para rechazar la hipótesis nula de que el porcentaje de la población con presión sistólica mayor o igual a 130 mm de Hg es igual a 0.30. El estadístico Z para la hipótesis es significativamente diferente de cero, y el p-valor es muy pequeño, lo que sugiere que la diferencia observada es altamente significativa.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

# Leer los datos
HIPERTENSION = pd.read_excel(r'C:\Users\lucia\Desktop\MASTER\ESTADISTICA\Documentación estadística-20231027\HIPERTENSION.xlsx')
grupo2 = HIPERTENSION[HIPERTENSION['Grupo'] == 2].copy()

grupo2['Diferencia_presion'] = grupo2['Presión sistólica después'] - grupo2['Presion sistólica antes']

statistic, p_valor = ttest_rel(grupo2['Presión sistólica después'], grupo2['Presion sistólica antes'])

print(f"Estadístico de prueba: {statistic}")
print(f"P-valor: {p_valor}")

plt.figure(figsize=(12, 6))

# Histograma
plt.subplot(1, 2, 1)
plt.hist(grupo2['Diferencia_presion'], bins='auto', color='darkorange', edgecolor='black')
plt.title('Histograma de Diferencias en Presión Sistólica')
plt.xlabel('Diferencia en Presión Sistólica')
plt.ylabel('Frecuencia')
plt.grid(axis='y', alpha=0.75)

# Gráfico de barras de las diferencias
plt.subplot(1, 2, 2)
plt.bar(range(len(grupo2)), grupo2['Diferencia_presion'], color='blue')
plt.xlabel('Individuo')
plt.ylabel('Diferencia de presión sistólica')
plt.title('Diferencia de presión sistólica antes y después del medicamento en el grupo 2')

# Ajustar el diseño
plt.tight_layout()

# Mostrar los gráficos
plt.show()

# Comparar el p-valor con el nivel de significación ( 0.05)
nivel_significacion = 0.05
if p_valor < nivel_significacion:
    print("Se rechaza la hipótesis nula. Hay evidencia significativa de que la presión sistólica cambia después de la toma del medicamento.")
else:
    print("No se puede rechazar la hipótesis nula. No hay evidencia significativa de que la presión sistólica cambie después de la toma del medicamento.")
```

El estudio prueba negativo sugiere que, en promedio, la presión sistólica después del medicamento es menor que la presión sistólica antes del medicamento en el Grupo 2. Con un p-valor inferior al nivel de significación del 5%, se rechaza la hipótesis nula, indicando evidencia significativa de que la presión sistólica cambia después de la toma del medicamento en adultos.
